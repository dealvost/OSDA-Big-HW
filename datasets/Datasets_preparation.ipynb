```python
heart_failure = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
heart_failure_ishodnik = heart_failure.copy()
```

```python

# Определение порогов и меток для бинаризации
bins_age = [0, 20, 40, 60, 80, float('inf')]
labels_age = ['<20', '20-39', '40-59', '60-79', '>=80']

bins_cpk = [0, 120, 250, float('inf')]
labels_cpk = ['Low_CPK', 'Medium_CPK', 'High_CPK']

bins_ef = [0, 40, 55, float('inf')]
labels_ef = ['Low_EF', 'Medium_EF', 'High_EF']

bins_platelets = [0, 150000, 450000, float('inf')]
labels_platelets = ['Low_Platelets', 'Normal_Platelets', 'High_Platelets']

bins_creatinine = [0, 0.7, 1.2, float('inf')]
labels_creatinine = ['Low_Creatinine', 'Normal_Creatinine', 'High_Creatinine']

bins_sodium = [0, 135, 145, float('inf')]
labels_sodium = ['Low_Sodium', 'Normal_Sodium', 'High_Sodium']

# Бинаризация времени (time) с использованием перцентилей
time_33 = heart_failure['time'].quantile(0.33)
time_66 = heart_failure['time'].quantile(0.66)

# Определение функции для бинаризации времени
def binarize_time(time):
    if time < time_33:
        return 'Short'
    elif time < time_66:
        return 'Medium'
    else:
        return 'Long'

# Бинаризация числовых переменных
heart_failure['age_group'] = pd.cut(heart_failure['age'], bins=bins_age, labels=labels_age, right=False)
heart_failure['creatinine_phosphokinase_group'] = pd.cut(heart_failure['creatinine_phosphokinase'], bins=bins_cpk, labels=labels_cpk, right=False)
heart_failure['ejection_fraction_group'] = pd.cut(heart_failure['ejection_fraction'], bins=bins_ef, labels=labels_ef, right=False)
heart_failure['platelets_group'] = pd.cut(heart_failure['platelets'], bins=bins_platelets, labels=labels_platelets, right=False)
heart_failure['serum_creatinine_group'] = pd.cut(heart_failure['serum_creatinine'], bins=bins_creatinine, labels=labels_creatinine, right=False)
heart_failure['serum_sodium_group'] = pd.cut(heart_failure['serum_sodium'], bins=bins_sodium, labels=labels_sodium, right=False)
heart_failure['time_group'] = heart_failure['time'].apply(binarize_time)

# Удаление исходных количественных столбцов
heart_failure = heart_failure.drop(['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium', 'time'], axis=1)

# Применение one-hot encoding и преобразование булевых значений в 0 и 1
heart_failure_encoded = pd.get_dummies(heart_failure).astype(int)

heart_failure = heart_failure_encoded



```
